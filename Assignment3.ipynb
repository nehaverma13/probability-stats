# problem 1

In a survey conducted by a non-banking financial company, a sample of 200 customers yielded that x of them were highly satisfied with the timely disbursal of their loans.
Write a Python code to perform the following operations:
1.Read an integer input thatspecifies the number of highly satisfied customers
2.Calculate an approximate 90% confidence interval for the proportion of the loan customers who are highly satisfied with disbursal time
•Find out the Margin of Error using scipy.stats.norm.ppf
•Calculate and print the confidence interval values rounded up to five decimal places and separated by a spaceNote:
•Margin of Error = Critical Value * Standard Error of Statistic
•Confidence Interval = Sample Statistic ± Margin of Error
Example: Let's say 172 out of 200 customers were highly satisfied.
Sample Input:172
The confidence interval should be printed as -
Sample Output:0.82856 0.89144

```python
#Read an integer input thatspecifies the number of highly satisfied customers
n = 200
x = int(input("Please Enter a number of happy customers"))                 
```

```python
p=x/n
p
```

```python
from scipy import stats
from math import sqrt
```

```python
stats.norm.ppf(0.95)  #90% confidence interval
```

```python
#critical error
critical_value = stats.norm.ppf(0.95)
critical_value
```

```python
#statistic error(se)
se = sqrt((p*(1-p))/n)       
se
```

```python
Marginal_Error = critical_value*se
Marginal_Error
```

#### Confidence Interval = Sample Statistic ± Margin of Error

```python
#calculate confidence interval
a=p+Marginal_Error
a
```

```python
#confidence interval values rounded up to five decimal places
np.around(a, 5)
```

```python
b=p-Marginal_Error
b
```

```python
#confidence interval values rounded up to five decimal places
np.around(b,5)
```

#### Calculate the probability of flood given it rained more than 500 mm in June (P(A|B))

P(F): Probability of flooding
P(J): Probability of having more than 500 mm rain in June
P(F ∩ J): Probability of flooding and having more than 500 mm rain in June
P(F|J): Probability of flooding given it rained more than 500 mm in June

```python
P_F = (6 + 54) / (6 + 54 + 19 + 39)
P_J = (39 + 54) / (6 + 54 + 19 + 39)
P_F_intersect_J = 54 / (6 + 54 + 19 + 39)
print(f"P(F): {P_F}") 
print(f"P(J): {P_J}")
print(f"P(F AND J): {P_F_intersect_J}")
```

```python
# Now calculate probailitity of flood given it rained more than 500 mm in June (P(A|B))
P_F_J = P_F_intersect_J / P_J
print(f"P(F|J): {P_F_J}")
```

```python
#Using the formula - P(A|B) = P(A ∩ B) / P(B) we can easily calculate condition probability
P_F_J=P_F_intersect_J/P_J
print(f"probailitity of flood given it rained more than 500 mm in June (P(A|B))\nP(FLOODS|JUNE):{P_F_J}")
```

### bayes formula
P(A∩B)=P(A|B).P(B)=P(B|A).P(A)
so,P(B|A)=P(A|B).P(B)/P(A)

```python
# Calculate the probability of rain more than 500 mm in June, given it flooded that year (P(B|A))
P_J_F=(P_F_J*P_J)/P_F
print(f" probability of rain more than 500 mm in June, given it flooded that year (P(B|A))\nP(JUNE|FLOODS):{P_J_F}" )
```

```python
from sklearn.datasets import load_wine
import pandas as pd
wine = load_wine()
df = pd.DataFrame(wine.data, columns = wine.feature_names)
df
```

```python
df.isnull().sum()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

#rmse = mean_squared_error(y_test,y_pred,squared=False)
```

```python
#Generate the sample size of 50 and give a randomstate as 100.
sample = df.sample(n=50,random_state=100)
```

```python
sample
```

```python
import math
from scipy import stats
z_critical = stats.norm.ppf(q=0.95) # Z_critical
```

```python
#Calculate Z-critical
z_critical
```

```python
pop_std = sample['alcohol'].std()
```

```python
pop_std
```

```python
sample_size = 50
margin_of_error = z_critical*(pop_std/math.sqrt(sample_size))
```

```python
#calculate margin of error
margin_of_error
```

```python
#calculate Confidence interval
sample_mean = sample['alcohol'].mean()
```

```python
sample_mean
```

```python
sample_mean + margin_of_error, sample_mean - margin_of_error
```
